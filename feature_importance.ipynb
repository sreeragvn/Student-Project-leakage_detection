```python
import pandas as pd
data = pd.read_csv('../data_acquisition/clean_data_check0.csv')
data = data.rename(columns={'number of leakage': 'number_of_leakage'})
data = data[(data.number_of_leakage == 1)]
# data= data.dropna()
data = data.drop(columns=['sample_number', 'x2','y2','Comments', 'Day', 'number_of_leakage', 'quality'])
y = data[['x1', 'y1']]
X = data.drop(['x1', 'y1'], axis=1)
```

```python
data.head()
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
# define dataset
# define the model
model = LinearRegression()
# fit the model
model.fit(X, y)
# get importance
importance = model.coef_
importance = importance[0]
# summarize feature importance
for i,v in enumerate(importance):
 print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```
